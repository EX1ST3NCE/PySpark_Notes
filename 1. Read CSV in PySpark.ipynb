### Load CSV
--------------

1. Go to Data
2. Create Table
3. Upload file
4. Create table in NoteBook

```
# File location and type
file_location = "/FileStore/tables/Data.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","


# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
# df.show()
```

>infer_schema

 *If infer_schema is false, it will treat all the fields as String.*

<br>

> first_row_is_header


*If first_row_is_header is false, it will treat the first row as a column in the table.*

<br>

```
df.printSchema()
```

```
df.createOrReplaceTempView("Data")
```
